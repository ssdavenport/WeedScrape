```python
from bs4 import BeautifulSoup
import urllib
import re
```

```python
# load XML file, import to BS object
token = "42791c5570b7b89fdc20f6eda3a181095dc681fcc2541791"
leafly1path = "/Users/sdavenpo/Google Drive (stevenseandavenport@gmail.com)/3. Dport Research/Projects/Weed Scrape/data/leafly dispensary-info-1 2.28.18.xml"
data = open(leafly1path, 'r').read()
soup = BeautifulSoup(data, 'html.parser')
```

```python
# Get URL nodes (list)
letters = soup.find_all("url") 
# Extract text for each element -> list
links = [ x.get_text() for x in letters]
```

```python
# Repeat for second half of sitemap
leafly2path = "/Users/sdavenpo/Google Drive (stevenseandavenport@gmail.com)/3. Dport Research/Projects/Weed Scrape/data/leafly dispensary-info-2 2.28.18.xml"
data2 = open(leafly2path, 'r').read()
soup2 = BeautifulSoup(data2, 'html.parser')
letters2 = soup2.find_all("url") 
links = [ x.get_text() for x in letters] + [ x.get_text() for x in letters2]
# Perform regex on URLs to narrow them down.
# I want to find anything after dispensary-info/.../ and remove it!
# TO remove redundant links, grab all of URL up to home page level.
# Get everything to the left of the first / after "info/".
links = [re.compile('(?=nfo)!n.*/.*/').sub('', link) for link in links]
```

```python
# Scrape a sample dispensary page
r = urllib.urlopen(links[2]).read()
soup = BeautifulSoup(r)

# Get divs with itemprop=='address'
address = soup.find('div', attrs={"itemprop" : "address"}).get_text()
join_time = soup.find("time").get_text()
reviews = soup.find('span', attrs={"itemprop" : "ratingCount"}).get_text()

dict = {"address":address,
        "joined":join_time,
        "reviews":reviews}
```

```python
dict_list = []

for link in links[:20]:
    
    soup = BeautifulSoup(urllib.urlopen(link).read())

    try: 
        address = soup.find('div', attrs={"itemprop" : "address"}).get_text()
    except:
        address = ""
    try: 
        join_time = soup.find("time").get_text()
    except:
        join_time = ""
    try: 
        reviews = soup.find('span', attrs={"itemprop" : "ratingCount"}).get_text()
    except:
        reviews = ""

    dict = {"url":link,
            "address":address,
            "joined":join_time,
            "reviews":reviews}
    
    dict_list.append(dict)
```

```python
# Turn that into a pandas data.frame
import pandas as pd
import csv
store_data = pd.DataFrame(dict_list)
store_data.to_csv("store_data.csv")
```
